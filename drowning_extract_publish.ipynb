```python
#this is the big extract block
# reads fixed width file, pull coded data from specific locations 
#File name was edited from the zip file for readability, link in read me
universal_set = pd.read_fwf('2023 mortality.fwf', colspecs = [*data_map[2023].values()], header=None, names=[*data_map[2023].keys()])
# preserving 4 digit icd code, for future analysis 
universal_set['icd_raw'] = universal_set['icd']
# slicing to 3 digit icd code
universal_set['icd'] = universal_set['icd'].str[:3]
# adding a year column for future analysis
universal_set['year'] = 2023
# filtering down to only drowning deaths via the trimmed icd code
drowning_test = universal_set[universal_set['icd'].isin(icd_10_drowning_codes.keys())]
# replacing coded values with the decode from my maps
drowning_test = drowning_test.replace(map_2023)
# sneak peak at my data to make sure decode was sucessful
drowning_test.head(20)
```

```python
#decoding the universal_set
universal_set = universal_set.replace(map_2023)
#checking decode
universal_set.head(20)
```
